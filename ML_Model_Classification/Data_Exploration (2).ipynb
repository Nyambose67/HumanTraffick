# Data Visualization

Count Plot of Labels

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='label', palette='Set2')
plt.title('Count of Phishing vs. Non-Phishing Emails')
plt.xlabel('Label (0 = Non-Phishing, 1 = Phishing)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Phishing', 'Phishing'])
plt.grid(axis='y')
plt.show()
```

**Word Cloud for Subject**:

The word clouds for the subject show the most common words, with more frequently occurring words appearing larger.

```python
plt.figure(figsize=(10, 6))
subject_text = ' '.join(df['subject'].dropna())
wordcloud_subject = WordCloud(width=800, height=400, background_color='white').generate(subject_text)
plt.imshow(wordcloud_subject, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Email Subjects', fontsize=20)
print('')
plt.show()
```

**Word Cloud for Body**:

The word clouds for the body show the most common words, with more frequently occurring words appearing larger.

```python
plt.figure(figsize=(10, 6))
body_text = ' '.join(df['body'].dropna())
wordcloud_body = WordCloud(width=800, height=400, background_color='white').generate(body_text)
plt.imshow(wordcloud_body, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Email Bodies', fontsize=20)
plt.show()
```
