Neccessary Imports

```python
import pandas as pd
import numpy as np
import re
```

Load the dataset

```python
df = pd.read_csv('/content/drive/MyDrive/CEAS_08.csv')
```

#Information about the dataset

```python
print("Dataset Shape:", df.shape)
print("Column Names:", df.columns)
print("\nData Types:\n", df.dtypes)
print("\nFirst 5 Rows:\n")
df.head()
```

 Check for missing values

```python
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values[missing_values > 0])
```

Handle missing values

```python
df = df.dropna(subset=['label'])#delete missing values on target variable
```

Convert date column to datetime format

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df.head()
```

Clean the text data in the subject and body columns

```python
def clean_text(text):
    if not isinstance(text, str):  # Check if the text is a string
        return ''
    # Remove URLs to focus the actually content
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove special characters and numbers
    text = re.sub(r'\W+', ' ', text)
    # Convert to lowercase make the ML model fast
    text = text.lower()
    return text

df['subject'] = df['subject'].apply(clean_text)
df['body'] = df['body'].apply(clean_text)
```

Save the cleaned DataFrame to a new CSV file

```python
cleaned_file_path = '/content/drive/MyDrive/cleaned_phishing_emails.csv'
df.to_csv(cleaned_file_path, index=False)
```
